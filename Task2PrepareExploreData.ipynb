# Limpieza de los datos

### Se elimina la columna ID

```python
# Se elimina la columna "ID" puesto que solo se trata de un contador de las filas
del credit['ID']
```

# Transformación de los datos

### Matriz de correlación

```python
corrMatriz = credit.corr()
print(corrMatriz)
f, ax = plt.subplots(figsize=(10, 8))
sea.heatmap(corrMatriz, mask=np.zeros_like(corrMatriz, dtype=np.bool), cmap=sea.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

### Matriz de varianza

```python
covMatriz = credit.cov()
print(covMatriz)
```

### Posterior al análisis de correlación y varianza

Se eliminan las columnas BILL_AMT2, BILL_AMT3, BILL_AMT4, BILL_AMT5 y BILL_AMT6 ya que se muestra una alta correlación entre ellas

```python
for columna in ['BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']:
    del credit[columna]
```

### Se asignan valores categóricos

```python
credit['SEX'] = credit['SEX'].map({1:'Masculino', 2:'Femenino'})
credit['EDUCATION'] = credit['EDUCATION'].map({0:'Otros', 1:'Graduate School', 2:'University', 3:'High School', 4:'Otros', 5:'Otros', 6:'Otros'})
credit['MARRIAGE'] = credit['MARRIAGE'].map({0:'Otros', 1:'Casado', 2:'Soltero', 3:'Divorciado'})
credit['default payment next month'] = credit['default payment next month'].map({0:'NO MOROSO', 1:'MOROSO'})
```

### Se convierten los datos de las columnas necesarias a categóricos

```python
for columna in ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'default payment next month']:
    credit[columna] = credit[columna].astype('category')
```

# Discretización de los datos

### La edad se discretiza en 6 grupos

```python
credit['AGE'] = pd.cut(credit['AGE'], 6, right = True, precision = 0)
```

### El monto de crédito en 10 grupos

```python
credit['LIMIT_BAL'] = pd.cut(credit['LIMIT_BAL'], 10, right = True, precision = 0)
```
